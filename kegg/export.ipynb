```r
library("KEGGREST") 
library("KEGGgraph")
```

```r
total_path.l<-unique(keggLink("pathway","sce"))
```

```r
path_id<-sapply(total_path.l,function(i) strsplit(i,split=":")[[1]][2]) 
```

```r
head(path_id)
```

```r
sapply(path_id,function(i){
    fn<-paste(i,".xml",sep="")
    gR<-parseKGML2Graph(fn,genesOnly=F,expandGenes=T)
    node_attr<-sapply(nodes(gR),function(i) strsplit(i,split=":")[[1]][1])
    node_id<-nodes(gR)
    idx<-which(node_attr%in%c("cpd","sce"))
    node.m<-cbind(node_id[idx],node_id[idx],node_attr[idx])
    colnames(node.m)<-c("Id","Label","Attr")
    out.fn<-paste(i,"_node.csv",sep = "",collapse = "")
    write.csv(node.m,out.fn,row.names=F)
})
```

```r
sapply(path_id,function(i){
    fn<-paste(i,".xml",sep="")
    gR<-parseKGML2Graph(fn,genesOnly=F,expandGenes=T)
    node_attr<-sapply(nodes(gR),function(i) strsplit(i,split=":")[[1]][1])
    node_id<-sapply(nodes(gR),function(i) strsplit(i,split=":")[[1]][2])
    idx<-which(node_attr%in%c("cpd","sce"))
    node.m<-cbind(node_id[idx],node_id[idx],node_attr[idx])
    edges.l<-edges(gR)
    edge.m<-Reduce(rbind,sapply(rownames(node.m),function(j){
        target.l<-edges.l[[j]]
        if(length(target.l)>0)
        cbind(rep(j,length(target.l)),target.l)
}))
    out.fn<-paste(i,"_gg_node.csv",sep = "",collapse = "")
    write.csv(edge.m,out.fn,row.names=F)
})
```
